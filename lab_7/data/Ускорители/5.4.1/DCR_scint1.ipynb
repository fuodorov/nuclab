```python
import array
from array import array
import matplotlib.pyplot as plt
```

```python
##Draw DCR

##set up the correct directory name with data
path = "C:/Users/Nuclab/Desktop/Ускорители/5.4.1"
##set up the correct file name with data
fileName1 = "5.4.1_no_tile_staircase.txt"
fileName2 = "5.4.1_with_tile_staircase.txt"
fileName3 = "5.4.1_with_tile_and_isotope_staircase.txt"
# fileName4 = "Sr_with_scint_53,92V_40db_gate100ms_2_staircase.txt"

listFiles = [fileName1, fileName2, fileName3] 

outFile = 'DCR_sipm.png'
    
for fileName in listFiles:
    file = open(path + "/" + fileName)
    trhU, freq = array( 'd' ), array( 'd' )
    for line in file.readlines():
        line = line.replace(',','.')
        b = line.split()
        if b:
            trhU.append(float(b[0]))
            freq.append(float(b[1]))
    
    plt.plot(trhU,freq, 'o', ls='-', ms=4)

plt.grid(True)
plt.yscale('log')
plt.xlabel('Threshold, mV')
plt.ylabel('Frequency, Hz')
plt.title('DCR')

plt.savefig(outFile )
plt.show()
```
